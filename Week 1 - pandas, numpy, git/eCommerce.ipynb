Part 2: Data Cleaning
• 2.1: Identify and handle missing values in the dataset. Options include dropping, filling with a placeholder, or imputing
based on other data points.

```python
import pandas as pd

#Part 1: Data Ingestion
df = pd.read_csv('ecommerce_sales.csv')

#Part 2: Data Cleaning
df.dropna(inplace=True, ignore_index=True)
df[10:21]
```

2.2: Ensure all data types are correct, focusing on converting order_date to a datetime format and checking numerical
columns for consistency.

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['product_id'] = df['product_id'].astype(int)
df['quantity'] = df['quantity'].astype(int)
df['customer_id'] = df['customer_id'].astype(int)
print(df.dtypes)
df.head(5)
```

Part 3: Data Manipulation

• 3.1: Create a new column total_price that represents the total amount spent on each order (calculated as quantity *
unit_price).

• 3.2: Extract additional features from the order_date column, such as order_weekday (the day of the week), and
order_month (month of the year).

```python
df['total_price'] = df['unit_price'] * df['quantity']
```
